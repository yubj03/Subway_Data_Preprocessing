```python
df2023 = pd.read_csv('서울교통공사_역별요일별환승인원_20231231.csv', encoding='cp949')
```

```python
line1 = pd.read_csv('서울교통공사 노선별 지하철역 정보.csv', encoding='cp949')
```

```python
line1
```

```python
df2023 = df2023[df2023['역명'].isin(line1['전철역명'])]
```

```python
df2019 = pd.read_csv('서울교통공사_역별요일별환승인원_20191231.csv', encoding='cp949')
```

```python
df2019
```

```python
df2019 = df2019[df2019['역명'].isin(line1['전철역명'])]
```

```python
df2019
```

```python
df2015 = pd.read_csv('서울교통공사_역별요일별환승인원_2015.csv', encoding='cp949')
```

```python
df2015
```

```python
df2015 = df2015[df2015['역 명'].isin(line1['전철역명'])]
```

```python
df2015
```

## 년도별 data columns 구분

```python
df2023 = df2023.rename(columns={'평일(일평균)': '2023평일(일평균)', '토요일': '2023토요일', '일요일': '2023일요일'})
```

```python
df2023
```

```python
df2019 = df2019.rename(columns={'평일(일평균)': '2019평일(일평균)', '토요일': '2019토요일', '일요일': '2019일요일'})
```

```python
df2019
```

```python
df2015 = df2015.rename(columns={'역 명': '역명', '평일(일평균)': '2015평일(일평균)', '토요일': '2015토요일', '일요일': '2015일요일'})
```

```python
df2015
```

## 데이터 병합

```python
pre_df = pd.merge(df2023, df2019, on='역명', how='outer')
df = pd.merge(pre_df, df2015, on='역명', how='outer')
```
