```python
df2019 = pd.read_csv('서울교통공사_지하철혼잡도정보_20191231.csv', encoding='cp949')
```

```python
df2019 = df2019[df2019['호선'] == '1호선']
```

```python
df2019['호선'] = df2019['호선'].replace('1호선', 1)
```

```python
df2019 = df2019.rename(columns={'역명': '출발역'})
```

```python
df2019
```

```python
df2017 = pd.read_csv('서울교통공사_지하철혼잡도정보_20171231.csv', encoding='cp949')
```

```python
df2017 = df2017[df2017['호선'] == '1']
```

```python
df2017['호선'] = df2017['호선'].astype(int)
```

```python
df2017 = df2017.rename(columns={'역명': '출발역'})
```

```python
df2017
```

```python
df2015 = pd.read_csv('서울교통공사_지하철혼잡도정보_2015.csv', encoding='cp949')
```

```python
df2015 = df2015[df2015['호선'] == '1호선']
```

```python
df2015['호선'] = df2015['호선'].replace('1호선', 1)
```

```python
df2015['역명'] = df2015['역명'].str.replace(r'\(\d+\)', '', regex=True)
```

```python
df2015 = df2015.rename(columns={'역명': '출발역'})
```

```python
df2015
```
