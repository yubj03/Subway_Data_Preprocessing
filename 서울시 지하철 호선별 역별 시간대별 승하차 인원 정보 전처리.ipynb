```python
import pandas as pd
```

```python
df = pd.read_csv('서울시 지하철 호선별 역별 시간대별 승하차 인원 정보.csv', encoding='cp949')
```

```python
#1호선 데이터 추출
df = df[df['호선명'] == '1호선']
```

```python
#2020~2022 데이터 제거
df = df[~((df['사용월']//100 >= 2020) & (df['사용월']//100 <= 2022))]
```

```python
df['호선명'] = df['호선명'].replace('1호선', 1)
```
