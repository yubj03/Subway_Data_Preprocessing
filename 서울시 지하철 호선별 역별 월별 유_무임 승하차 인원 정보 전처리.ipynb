```python
import pandas as pd
```

```python
df = pd.read_csv('서울시 지하철 호선별 역별 월별 유_무임 승하차 인원 정보.csv', encoding='cp949')
```

```python
#1호선 데이터 추출
df = df[df['호선명'] == '1호선']
```

```python
#2020~2022 데이터 제거
df = df[~((df['사용월']//100 >= 2020) & (df['사용월']//100 <= 2022))]
```

```python
df
```

```python
#승차인원 산출(유임+무임)
df['승차인원'] = df['유임승차인원'] + df['무임승차인원']
df.drop(columns=['유임승차인원', '무임승차인원'], inplace=True)
```

```python
#하차인원 산출(유임+무임)
df['하차인원'] = df['유임하차인원'] + df['무임하차인원']
df.drop(columns=['유임하차인원', '무임하차인원'], inplace=True)
```

```python
df['호선명'] = df['호선명'].replace('1호선', 1)
```
